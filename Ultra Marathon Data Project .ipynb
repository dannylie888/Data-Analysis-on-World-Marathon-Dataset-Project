```python
#Combine all filters together
df[(df['Event distance/length'].isin(['50mi','50km'])) & ((df['Year of event'] == 2020) & (df['Event name'].str.split('(').str.get(1).str.split(')').str.get(0) == 'USA'))]
```

```python
dffiltered = df[(df['Event distance/length'].isin(['50mi','50km'])) & ((df['Year of event'] == 2020) & (df['Event name'].str.split('(').str.get(1).str.split(')').str.get(0) == 'USA'))]
```

```python
dffiltered.shape
```

```python
#remove (USA) from event name
dffiltered['Event name'] = dffiltered['Event name'].str.split('(').str.get(0)
```

```python
dffiltered.head()
```

```python
#Clean up athlete age
dffiltered['Athlete_age'] = dffiltered['Year of event'] - dffiltered['Athlete year of birth'] 
```

```python
#remove h from athlete performance
dffiltered['Athlete performance']=dffiltered['Athlete performance'].str.split(' ').str.get(0)
```

```python
dffiltered.head(5)
```

```python
#Drop column ( Athlete club, Athlete country, Athelete year of birth, Athelete Age category)
dffiltered = dffiltered.drop(['Athlete club', 'Athlete country', 'Athlete year of birth', 'Athlete age category'], axis=1)
```

```python
dffiltered.head()
```

```python
#Clean up null values
dffiltered.isna().sum()
```

```python
dffiltered[dffiltered['Athlete_age'].isna()==1]
```

```python
dffiltered = dffiltered.dropna()
```

```python
dffiltered.shape
```

```python
#Check Dupes
dffiltered[dffiltered.duplicated() == True]
```

```python
#Reset index
dffiltered.reset_index(drop = True)
```

```python
#Fix types
dffiltered.dtypes
dffiltered['Athlete_age'] = dffiltered['Athlete_age'].astype(int)
```

```python
dffiltered.dtypes
```

```python
dffiltered['Athlete average speed'] = dffiltered['Athlete average speed'].astype(float)
```

```python
dffiltered.head()
```

```python
#Rename columns
#Year of event                 int64
#Event dates                  object
#Event name                   object
#Event distance/length        object
#Event number of finishers     int64
#Athlete performance          object
#Athlete gender               object
#Athlete average speed        object
#Athlete ID                    int64
#Athlete_age                   int32
```

```python
dffiltered = dffiltered.rename(columns ={'Year of event': 'year',
                                         'Event dates': 'race_day',
                                         'Event name': 'race_name',
                                         'Event distance/length': 'race_length',
                                         'Event number of finishers': 'race_number_of_finishers',
                                         'Athlete performance': 'athlete_performance',
                                         'Athlete gender': 'athlete_gender',
                                         'Athlete average speed': 'athlete_average_speed',
                                         'Athlete ID': 'athlete_id',
                                         'Athlete_age': 'athlete_age'    
})
```

```python
dffiltered.head()
```

```python
df3 = dffiltered[['race_day', 'race_name','race_length','race_number_of_finishers','athlete_id','athlete_gender','athlete_age','athlete_performance','athlete_average_speed']]
```

```python
#Difference in speed for the 50k, 50mi, male to female
df3.groupby(['race_length', 'athlete_gender'])['athlete_average_speed'].mean()
```

```python
#what age groups are the best in the 50m race(20+races min)
df3.query('race_length == "50mi"').groupby('athlete_age')['athlete_average_speed'].agg(['mean','count']).sort_values('mean',ascending = False).query('count>19')
```

```python
#what age groups are the worst in the 50m race(10+races min) (show 20)
df3.query('race_length == "50mi"').groupby('athlete_age')['athlete_average_speed'].agg(['mean','count']).sort_values('mean',ascending = True).query('count>9')
```

```python
#Seasons for the data -> Slower in summer than winter ?

#spring 3-5
#Summer 6-8
#fall 9-11
#winter 12-2

#split between to decimals
```

```python
df3['race_month'] = df3['race_day'].str.split('.').str.get(1).astype(int)
```

```python
df3['race_season'] = df3['race_month'].apply(lambda x: 'Winter' if x>11 else 'Fall' if x>8 else 'Summer' if x>5 else 'Spring' if x>2 else 'Winter')
```
